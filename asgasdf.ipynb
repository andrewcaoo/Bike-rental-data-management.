
# III. Cleaning data.

```python
# It not true. Because it assigned a new series of different length to the dataframe. 
# citibikes_df['Birth Year'] = citibikes_df[citibikes_df['Birth Year'].notna()]['Birth Year'].astype(int)
# citibikes_df['Birth Year'] = citibikes_df['Birth Year'].apply(lambda x: int(x) if not pd.isna(x) else x)

citibikes_df['Birth Year'] = citibikes_df['Birth Year'].astype('Int64')

citibikes_df.fillna(
    value={
        'User Type': 'Customer',
        'Birth Year': citibikes_df['Birth Year'].median()
    },
    inplace = True
)

citibikes_df['Start Time'] = pd.to_datetime(citibikes_df['Start Time'])
citibikes_df['Stop Time'] = pd.to_datetime(citibikes_df['Stop Time'])
```

```python
weather_data_df.drop(['TSUN','PGTM','STATION','NAME'], axis=1, inplace=True)

weather_data_df.fillna(
    value = {
        'WSF5': weather_data_df['WSF5'].median(),
        'WDF5': weather_data_df['WDF5'].median(),
    },
    inplace = True
) 

weather_data_df =  weather_data_df.astype({'DATE': 'datetime64[ns]'})

weather_data_df.head()
```

# V.Create database

## 1.Connection establishment

```python
from sqlalchemy import create_engine, text
from dotenv import dotenv_values
from sqlalchemy.engine import URL

config = dotenv_values(".env") 

url_object = URL.create(
    config['POSTGRES_ASYNC_PREFIX'],
    username=config['POSTGRES_USER'],
    password=config['POSTGRES_PASSWORD'],
    host=config['POSTGRES_SERVER'],
    database=config['POSTGRES_DB']
)

engine = create_engine(url_object)
print(engine)
```

## 2.Create the database

```python
# Delete the existing table.

drop_tables = """
DROP TABLE IF EXISTS Dim_Weather_Data;
DROP TABLE IF EXISTS Dim_Station;
DROP TABLE IF EXISTS Dim_Bike;
DROP TABLE IF EXISTS Dim_User;
DROP TABLE IF EXISTS Fact_Trip;
"""


create_table = """
CREATE TABLE Dim_Station (
    Station_id SERIAL PRIMARY KEY,
    Station_name VARCHAR(255),
    Station_latitude FLOAT,
    Station_longitude FLOAT
);

CREATE TABLE Dim_Bike (
    Bike_id SERIAL PRIMARY KEY
);

CREATE TABLE Dim_User (
    User_id SERIAL PRIMARY KEY,
    User_Type VARCHAR(255),
    Birth_Year INTEGER,
    Gender VARCHAR(10)
);

CREATE TABLE Fact_Trip (
    Trip_id SERIAL PRIMARY KEY,
    Trip_duration INT,
    Start_date DATE,
    Stop_date DATE,
    Start_station VARCHAR(100),
    Stop_station VARCHAR(100),
    Bike_id VARCHAR(100),
    Start_time TIME,
    Stop_time TIME,
    User_id SERIAL
);

CREATE TABLE Dim_Weather_Data (
    Date DATE PRIMARY KEY,
    AWND FLOAT,
    PRCP FLOAT,
    SNOW FLOAT,
    SNWD FLOAT,
    TAVG FLOAT,
    TMAX FLOAT,
    TMIN FLOAT,
    WDF2 FLOAT,
    WDF5 FLOAT,
    WSF2 FLOAT,
    WSF5 FLOAT
);

"""

with engine.connect() as connection:
    connection.execute(text(drop_tables))
    connection.execute(text(create_table))
```

## 3.Insert data from dataframe to the database

```python
# Insert data into Dim_bike table

citibikes_df[citibikes_df['Birth Year'].isna()]

Bike_df =  pd.DataFrame(citibikes_df['Bike ID']).drop_duplicates()

Bike_df.columns = ['bike_id']

Bike_df.to_sql('dim_bike', engine, if_exists='append', index=False)

# Insert data into Dim_User table.

user_df = citibikes_df[['User Type','Birth Year', 'Gender']].drop_duplicates()
user_df['user_id'] = range(1, len(user_df) + 1)

user_df.rename(columns={
        'User Type': 'user_type',
        'Birth Year': 'birth_year',
        'Gender': 'gender'
            }, inplace=True)
user_df.columns
user_df.to_sql('dim_user', engine, if_exists='append', index=False)

# Insert data into Dim_bike table Dim_Station

start_station_df =  citibikes_df[['Start Station ID', 'Start Station Name', 'Start Station Latitude','Start Station Longitude']].drop_duplicates()
end_station_df =  citibikes_df[['End Station ID', 'End Station Name', 'End Station Latitude','End Station Longitude']].drop_duplicates()
end_station_df.rename(columns={
    'End Station ID': 'Start Station ID', 
    'End Station Name': 'Start Station Name',
    'End Station Latitude': 'Start Station Latitude',
    'End Station Longitude': 'Start Station Longitude'
}, inplace=True)

station_df = pd.concat([end_station_df, start_station_df]).drop_duplicates()

station_df.rename(columns={
    'Start Station ID': 'station_id',
    'Start Station Name': 'station_name',
    'Start Station Latitude': 'station_latitude',
    'Start Station Longitude': 'station_longitude'
}, inplace=True)

station_df.to_sql('dim_station', engine, if_exists='append', index=False)

# Insert data into Dim_bike table dim_weather_data 
weather_df = weather_data_df.drop_duplicates()

weather_df.rename(columns={
    'DATE': 'date',
    'AWND': 'awnd',
    'PRCP': 'prcp',
    'SNOW': 'snow',
    'SNWD': 'snwd',
    'TAVG': 'tavg',
    'TMAX': 'tmax',
    'TMIN': 'tmin',
    'WDF2': 'wdf2',
    'WDF5': 'wdf5',
    'WSF2': 'wsf2',
    'WSF5': 'wsf5'
}, inplace= True)
weather_df.to_sql('dim_weather_data', engine, if_exists='append', index=False)

# Insert data into Dim_bike table Fact_trip
trip_df = citibikes_df[['Trip Duration','Start Date', 'Stop Date', 'Start Time', 'Stop Time', 'Start Station ID', 'End Station ID', 'Bike ID', 'User Type', 'Birth Year', 'Gender']]
trip_df.rename(columns={
    'Trip Duration': 'trip_duration',
    'Start Date': 'start_date',
    'Stop Date': 'stop_date',
    'Start Time': 'start_time',
    'Stop Time': 'stop_time',
    'Start Station ID': 'start_station',
    'End Station ID': 'stop_station',
    'Bike ID': 'bike_id'
    }, inplace=True)

final_trip_df = pd.merge(
                    left = trip_df,
                    right = user_df,
                    left_on  = ['User Type', 'Birth Year', 'Gender'],
                    right_on = ['user_type','birth_year','gender'],
                    how = 'inner'
)
final_trip_df.drop(['User Type', 'Birth Year', 'Gender', 'user_type','birth_year','gender'], axis = 1, inplace=True)
final_trip_df['trip_id'] = range(1, len(final_trip_df) + 1)
final_trip_df.to_sql('fact_trip', engine, if_exists='append', index=False)
```
