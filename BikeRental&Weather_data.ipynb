# I: Load the data

```python
path = 'bike-rental-starter-kit/bike-rental-starter-kit/data/'
weather_data_df = pd.read_csv(path + 'newark_airport_2016.csv')
weather_data_df
```

```python
files = glob.glob(path + "*citibike-tripdata.csv")

df_list = []
for f in (files):
    df = pd.read_csv(f)
    df_list.append(df)

citibikes_df = pd.concat(df_list)

citibikes_df.columns
```


# III) Cleaning data.

```python
# It not true. Because it assigned a new series of different length to the dataframe. 
# citibikes_df['Birth Year'] = citibikes_df[citibikes_df['Birth Year'].notna()]['Birth Year'].astype(int)
# citibikes_df['Birth Year'] = citibikes_df['Birth Year'].apply(lambda x: int(x) if not pd.isna(x) else x)

citibikes_df['Birth Year'] = citibikes_df['Birth Year'].astype('Int64')

citibikes_df.fillna(
    value={
        'User Type': 'Customer',
        'Birth Year': citibikes_df['Birth Year'].median()
    }
)

citibikes_df['Start Time'] = pd.to_datetime(citibikes_df['Start Time'])
citibikes_df['Stop Time'] = pd.to_datetime(citibikes_df['Stop Time'])
```

```python
# weather_data_df.drop(['TSUN','PGTM'], axis=1, inplace=True)

weather_data_df.fillna(
    value = {
        'WSF5': weather_data_df['WSF5'].median(),
        'WDF5': weather_data_df['WDF5'].median(),
    }
) 

weather_data_df =  weather_data_df.astype({'DATE': 'datetime64[ns]'})

weather_data_df.head()
```
